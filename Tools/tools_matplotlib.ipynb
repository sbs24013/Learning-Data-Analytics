```python
# 처음으로 그려보는 그래프

# matplotlib 라이브러리 임포트
import matplotlib

matplotlib.rc('font', family='NanumGothic')
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
import matplotlib.pyplot as plt

plt.plot([1,2,4,9,5,3])
plt.show()
```

```python
# 단일 배열: 수직 축의 좌표로서 이를 사용하게 되며, 각 데이터의 배열상 색인(인덱스)를 수평 좌표로 사용한다
# 두 개의 배열: 하나는 x축에 대한 것, 다른 하나는 y축에 대한 것

plt.plot([-3, -2, 5, 0], [1, 6, 4, 3]) 
```

```python
# 축은 자동으로 데이터에 가능한 마주처진다
# 약간의 공간을 더 주고 싶은 경우 axis 함수를 호출하여 각 축의 범위 [xmin, xmax, ymin, ymax] 를 확장 할 수 있다
plt.plot([-3, -2, 5, 0], [1, 6, 4, 3])
plt.axis([-4, 6, 0, 7]) # [xmin, xmax, ymin, ymax]
plt.show()
```

```python
# 수학 함수
# Numpy.linespace
# -2~2 범위에 속하는 500개의 부동 소수로 수성된 x 배열을 생성
# x의 각 값의 거듭제곱된 값을 포함하는 y 배열 생성

import numpy as np

x = np.linspace(-2, 2, 500)
y = x**2

plt.plot(x, y)
plt.show()
```

```python
# 타이틀, x 및 y 라벨, 모눈자 추가
plt.plot(x, y)
plt.title("2차 함수")
plt.xlabel("x")
plt.ylabel("y = x**2")
plt.grid(True)
plt.show()
```

```python
# 선의 스타일과 색상
# 기본적으로 맷플롭립은 바로 다음에 위치한(연이은) 데이터 사이에 선을 그린다
plt.plot([0, 100, 100, 0, 0, 100, 50, 0, 100], [0, 0, 100, 100, 0, 100, 130, 100, 0])
plt.axis([-10, 110, -10, 140])
plt.show()
```

```python
# 세번째 파라미터를 지정하면 선의 스타일과 색상을 바꿀 수 있다
# ex. 'g--': 초록색 파선

plt.plot([0, 100, 100, 0, 0, 100, 50, 0, 100], [0, 0, 100, 100, 0, 100, 130, 100, 0], "g--")
plt.axis([-10, 110, -10, 140])
plt.show()
```

```python
# x1, y1, [스타일1], x2, y2, [스타일2], ...
# 하나의 그래프에 여러개의 선과 각 선의 스타일을 간단히 그려볼 수 있다

plt.plot([0, 100, 100, 0, 0], [0, 0, 100, 100, 0], "r-", 
         [0, 100, 50, 0, 100], [0, 100, 130, 100, 0], "g--")
plt.axis([-10, 110, -10, 140])
plt.show()
```

```python
# show를 호출하기 전 plot을 여러번 호출할 수 있다

plt.plot([0, 100, 100, 0, 0], [0, 0, 100, 100, 0], "r-")
plt.plot([0, 100, 50, 0, 100], [0, 100, 130, 100, 0], "g--")
plt.axis([-10, 110, -10, 140])
plt.show()
```

```python
# 선 대신 간단한 점을 그려볼 수도 있다
# g-- 초록색 파선
# r: 빤간 점선
# b^: 파랑 삼각형

x = np.linspace(-1.4, 1.4, 30)
plt.plot(x, x, 'g--', x, x**2, 'r:', x, x**3, 'b^')
plt.show()
```

```python
# plot 함수는 Line2D 객체로 구성된 리스트를 반환한다
# 이 선들에 대한 추가적인 속성을 설정할 수도 있다
# ex. 선의 두께, 스타일, 투명도..

x = np.linspace(-1.4, 1.4, 30)
line1, line2, line3 = plt.plot(x, x, 'g--', x, x**2, 'r:', x, x**3, 'b^') # x, y, styles
line1.set_linewidth(3.0)
line1.set_dash_capstyle("round")
line3.set_alpha(0.2)
plt.show()
```

```python
# 그림의 저장
# 그래프를 그림파일로 저장하는 방법은 파일 이름을 지정하여 savefig() 함수를 호출해 주기만 하면 된다
# 가능한 이미지 포맷은 그래픽 백엔드에 따라서 지원 여부가 결정된다

x = np.linspace(-1.4, 1.4, 30)
plt.plot(x, x**2)
plt.savefig("my_square_function.png", transparent=True)
```

```python
# 부분 그래프 (subplot)

# 맷플롭립은 하나의 그림(figure)에 여러개의 부분 그래프를 포함할 수 있다
# 이 부분 그래프는 격자 형식으로 관리된다

# subplot 함수를 호출하여 그래프를 생성할 수 있다
# 이 때 격자의 행/열의 수 및 그래프를 그리고자 하는 부분 그래프의 색인을 파라미터로 지정해줄 수 있다
# 색인은 1부터 시작하여, 좌->우, 상단->하단의 방향이다

# pyplot은 현재 활성화된 부분 그래프를 계속해서 추적한다
# plt.gca()를 호출하여 해당 부분 그래프의 참조를 얻을 수 있다
# 따라서 plot 함수를 호출할 때 활성화된 부분 그래프에 그림이 그려지게 된다

x = np.linspace(-1.4, 1.4, 30)
plt.subplot(2, 2, 1)  # 2 행 2 열 크기의 격자 중 첫 번째 부분 그래프 = 좌측 상단
plt.plot(x, x)
plt.subplot(2, 2, 2)  # 2 행 2 열 크기의 격자 중 두 번째 부분 그래프 = 우측 상단
plt.plot(x, x**2)
plt.subplot(2, 2, 3)  # 2 행 2 열 크기의 격자 중 세 번째 부분 그래프 = 좌측 하단
plt.plot(x, x**3)
plt.subplot(2, 2, 4)  # 2 행 2 열 크기의 격자 중 네 번째 부분 그래프 = 우측 하단
plt.plot(x, x**4)
plt.show()

# subplot(223) = sunplot(2, 2, 3)
```

```python
# 격자의 여러 영역으로 확장된 부분 그래프를 생성하는 것도 쉽다

plt.subplot(2, 2, 1)  # 2 행 2 열 크기의 격자 중 첫 번째 부분 그래프 = 좌측 상단
plt.plot(x, x)
plt.subplot(2, 2, 2)  # 2 행 2 열 크기의 격자 중 두 번째 부분 그래프 = 우측 상단
plt.plot(x, x**2)
plt.subplot(2, 1, 2)  # 2행 *1* 열의 두 번째 부분 그래프 = 하단
plt.plot(x, x**3)
plt.show()
```

```python
# 보다 복잡한 부분 그래프의 위치 선정이 필요하다면, subplot2grid를 대신 사용할 수 있다
# 격자의 행과 열의 번호 및 격자에서 해당 부분 그래프를 그릴 위치를 지정해줄 수 있다
# 좌측상단 = (0,0)
# 몇 개의 행/열로 확장되어야 하는지도 추가적으로 지정할 수 있다

plt.subplot2grid((3,3), (0, 0), rowspan=2, colspan=2)
plt.plot(x, x**2)
plt.subplot2grid((3,3), (0, 2))
plt.plot(x, x**3)
plt.subplot2grid((3,3), (1, 2), rowspan=2)
plt.plot(x, x**4)
plt.subplot2grid((3,3), (2, 0), colspan=2)
plt.plot(x, x**5)
plt.show()
```

```python
# 여러개의 그림
# 각 그림은 하나 이상의 부분 그래프를 가질 수 있다
# 기본적으로 matplotlib이 자동으로 figure(1)을 생성한다
# 그림간 전활을 할 때, pyplot은 현재 활성화된 그림을 계속 추적한다
# plt.gcf()의 호출로 알 수 있다
# 활성화된 그림의 활성화된 부분 그래프가 현재 그래프가 그려질 부분 그래프가 된다

x = np.linspace(-1.4, 1.4, 30)

plt.figure(1)
plt.subplot(211)
plt.plot(x, x**2)
plt.title("2차식과 3차식")
plt.subplot(212)
plt.plot(x, x**3)

plt.figure(2, figsize=(10, 5))
plt.subplot(121)
plt.plot(x, x**4)
plt.title("y = x**4")
plt.subplot(122)
plt.plot(x, x**5)
plt.title("y = x**5")

plt.figure(1)      # 그림 1로 돌아가며, 활성화된 부분 그래프는 212 (하단)이 됩니다
plt.plot(x, -x**3, "r:")

plt.show()
```

```python
# Pyplot은 상태 머신을 완전히 무시할 수 있도록 지원한다 (명시적으로 작성 가능)
# subplots 함수를 호출해서 반환되는 figure 객체 및 축의 리스트를 사용

x = np.linspace(-2, 2, 200)
fig1, (ax_top, ax_bottom) = plt.subplots(2, 1, sharex=True)
fig1.set_size_inches(10,5)
line1, line2 = ax_top.plot(x, np.sin(3*x**2), "r-", x, np.cos(5*x**2), "b-")
line3, = ax_bottom.plot(x, np.sin(3*x), "r-")
ax_top.grid(True)

fig2, ax = plt.subplots(1, 1)
ax.plot(x, x**2)
plt.show()
```

```python
# 텍스트 그리기
# text 함수를 호출하여 텍스트를 그래프의 원하는 위치에 추가할 수 있다
# 출력을 원하는 텍스트와 수평 및 수직 좌표를 지정하고, 추가적으로 몇 가지 속성을 지정해 주면 된다
# 맷플롯립의 모든 텍스트는 TeX 방정식 표현을 포함할 수 있다

x = np.linspace(-1.5, 1.5, 30)
px = 0.8
py = px**2

plt.plot(x, x**2, "b-", px, py, "ro")

plt.text(0, 1.5, "2차 함수\n$y = x^2$", fontsize=20, color='blue', 
         horizontalalignment="center")
plt.text(px - 0.08, py, "어떤 포인트", ha="right", weight="heavy")
plt.text(px + 0.05, py - 0.4, "x = %0.2f\ny = %0.2f"%(px, py), rotation=-30, 
         color='gray')

plt.show()
```

```python
# ha = horizontal_alignment (수평정렬)

# annotate 함수: 그래프의 요소에 주석을 다는 것
# 관심있는 부분의 위치를 지정하고, 텍스트의 위치를 지정한다
# 텍스트 및 화살표에 대한 추가적인 속성도 지정해줄 수 있다

plt.plot(x, x**2, px, py, "ro")
plt.annotate("어떤 포인트", xy=(px, py), xytext=(px-1.3,py+0.5),
                           color="green", weight="heavy", fontsize=14,
                           arrowprops={"facecolor": "lightgreen"})
plt.show()
```

```python
# bbox 속성을 사용하면, 텍스트를 포함하는 사각형을 그려볼 수도 있다

plt.plot(x, x**2, px, py, "ro")

bbox_props = dict(boxstyle="rarrow,pad=0.3", ec="b", lw=2, fc="lightblue")
plt.text(px-0.2, py, "어떤 포인트", bbox=bbox_props, ha="right")

bbox_props = dict(boxstyle="round4,pad=1,rounding_size=0.2", ec="black", 
                  fc="#EEEEFF", lw=5)
plt.text(0, 1.5, "2차 함수\n$y = x^2$", fontsize=20, color='black', 
         ha="center", bbox=bbox_props)

plt.show()
```

```python
# 재미를 위해서 xkcd 스타일의 그래프를 그려보고 싶다면, with plt.xkcd() 섹션 블록을 활용할 수도 있습니다

with plt.xkcd():
  plt.plot(x, x**2, px, py, "ro")

  bbox_props = dict(boxstyle="rarrow,pad=0.3", ec="b", lw=2, fc="lightblue")
  plt.text(px-0.2, py, "어떤 포인트", bbox=bbox_props, ha="right")

  bbox_props = dict(boxstyle="round4,pad=1,rounding_size=0.2", ec="black", 
                      fc="#EEEEFF", lw=5)
  plt.text(0, 1.5, "2차 함수\n$y = x^2$", fontsize=20, color='black', 
    ha="center", bbox=bbox_props)

  plt.show()
```

```python
# 범례 (Legends)
# 범례를 추가하는 가장 간단한 방법은 모든 선에 라벨을 설정 해 주고, legend 함수를 호출하는 것이다

x = np.linspace(-1.4, 1.4, 50)
plt.plot(x, x**2, "r--", label="2차 함수")
plt.plot(x, x**3, "g-", label="3차 함수")
plt.legend(loc="best")
plt.grid(True)
plt.show()
```

```python
# 비선형 척도
# Matplotlib은 로그, 로짓과 같은 비선형 척도를 지원한다

x = np.linspace(0.1, 15, 500)
y = x**3/np.exp(2*x)

plt.figure(1)
plt.plot(x, y)
plt.yscale('linear')
plt.title('linear')
plt.grid(True)

plt.figure(2)
plt.plot(x, y)
plt.yticks(fontname="Liberation Sans")
plt.yscale('log')
plt.title('log')
plt.grid(True)

plt.figure(3)
plt.plot(x, y)
plt.yticks(fontname="Liberation Sans")
plt.yscale('logit')
plt.title('logit')
plt.grid(True)

plt.figure(4)
plt.plot(x, y - y.mean())
plt.yticks(fontname="Liberation Sans")
plt.yscale('symlog', linthresh=0.05)
plt.title('symlog')
plt.grid(True)

plt.show()
```

```python
# 틱과 틱커 (Ticks and tickers)
# "틱"은 표시의 위치이며
# "틱 선"은 그 위치에 그려지는 작은 선이다
# "틱 라벨"은 틱 선 옆에 그려지는 라벨이며
# "틱커"는 틱의 위치를 결정하는 객체이다

x = np.linspace(-2, 2, 100)

plt.figure(1, figsize=(15,10))
plt.subplot(131)
plt.plot(x, x**3)
plt.grid(True)
plt.title("기본 틱")

ax = plt.subplot(132)
plt.plot(x, x**3)
ax.xaxis.set_ticks(np.arange(-2, 2, 1))
plt.grid(True)
plt.title("x 축에 수동으로 틱 설정하기")

ax = plt.subplot(133)
plt.plot(x, x**3)
plt.minorticks_on()
ax.tick_params(axis='x', which='minor', bottom='off')
ax.xaxis.set_ticks([-2, 0, 1, 2])
ax.yaxis.set_ticks(np.arange(-5, 5, 1))
ax.yaxis.set_ticklabels(["min", -4, -3, -2, -1, 0, 1, 2, 3, "max"])
plt.grid(True)
plt.title("y 축에 수동으로 틱과 레이블을\n설정하기(마이너 틱 추가)")

plt.show()
```

```python
# 극좌표계의 투영 (Polar projection)
# 극좌표계 그래프를 그리는 것은 매우 간단하다
# 부분 그래프를 생성할 때 projection 속성을 "polar"로 설정해 주기만 하면 된다

radius = 1
theta = np.linspace(0, 2*np.pi*radius, 1000)

plt.subplot(111, projection='polar')
plt.plot(theta, np.sin(5*theta), "g-")
plt.plot(theta, 0.5*np.cos(20*theta), "b-")
plt.show()
```

```python
# 3차원 투열
# 부분 그래프를 만들 때 projection 속성을 '3d'로 설정하면 된다 (3D 축 객체를 반환)
# 이 객체의 plot_surface 메서드를 호출 할 때, x, y, z 좌표를 포함한 추가적인 속성을 지정할 수 있다

x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

figure = plt.figure(1, figsize = (12, 4))
subplot3d = plt.subplot(111, projection='3d')
surface = subplot3d.plot_surface(X, Y, Z, rstride=1, cstride=1,
                                 cmap=matplotlib.cm.coolwarm, linewidth=0.1)
plt.show()
```

```python
# 동일한 데이터를 출력하는 또 다른 방법은 등고선도(contour plot)를 이용

plt.contourf(X, Y, Z, cmap=matplotlib.cm.coolwarm)
plt.colorbar()
plt.show()
```

```python
# 산점도(Scatter plot)
# 단순히 각 점에 대한 x 및 y 좌표를 제공하면 산점도를 그릴 수 있다

np.random.seed(42)  # 노트북 결과를 일정하게 만들기 위해

x, y = np.random.rand(2, 100)
plt.scatter(x, y)
plt.show()
```

```python
# 부수적으로 각 점의 크기를 정할 수도 있다

x, y, scale = np.random.rand(3, 100)
scale = 500 * scale ** 5
plt.scatter(x, y, s=scale)
plt.show()
```

```python
# 여러 속성 지정 가능
# 테두리 및 모양의 내부 색상, 그리고 투명도와 같은것

for color in ['red', 'green', 'blue']:
  n = 100
  x, y = np.random.rand(2, n)
  scale = 500.0 * np.random.rand(n) ** 5
  plt.scatter(x, y, s=scale, c=color, alpha=0.3, edgecolors='blue')

plt.grid(True)

plt.show()
```

```python
# 선
# plot 함수를 사용하여 선을 그릴 수 있다
# 그래프를 통과하는 무한한 선을 그리는 유틸리티 함수를 만들면 편리하다 (기울기와 절편사용)
# hlines 및 vlines 함수를 사용하면, 아래와 같이 부분 수평 및 수직 선을 그릴 수도 있다

def plot_line(axis, slope, intercept, **kargs):
    xmin, xmax = axis.get_xlim()
    plt.plot([xmin, xmax], 
             [xmin*slope+intercept, xmax*slope+intercept], 
             **kargs)

x = np.random.randn(1000)
y = 0.5*x + 5 + np.random.randn(1000)*2
plt.axis([-2.5, 2.5, -5, 15])
plt.scatter(x, y, alpha=0.2)
plt.plot(1, 0, "ro")
plt.vlines(1, -5, 0, color="red")
plt.hlines(0, -2.5, 1, color="red")
plot_line(axis=plt.gca(), slope=0.5, intercept=5, color="magenta")
plt.grid(True)
plt.show()
```
